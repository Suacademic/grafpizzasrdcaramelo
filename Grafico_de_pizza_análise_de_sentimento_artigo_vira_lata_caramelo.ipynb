```python
import matplotlib.pyplot as plt

#Dados de sentimentos relacionados ao artigo https://www.petlove.com.br/dicas/cachorro-vira-lata-caramelo
emotion = ['sadness', 'joy', 'fear', 'disgust', 'anger']
quantidades = [0.27266421568627452, 0.48294009515570935, 0.3146875, 0.18039485294117646, 0.15588235294117647] # Em porcentagem

# Plotar o gráfico de pizza
plt.pie(quantidades, labels=emotion, autopct='%1.1f%%', startangle=140)

# Personalização do gráfico
plt.title('Grafico de emoções relacionado ao artigo do vira lata caramelo')

# Mostrar o gráfico
plt.savefig ("grafiv.png")
```
